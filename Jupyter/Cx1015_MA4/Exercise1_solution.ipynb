```python
df = pd.read_csv('../Resources/adult.data', header = None)
df.head(1)
```

```python
dictionary = {}

for year in range(2000,2017,4):
    df = pd.read_html(f'https://en.wikipedia.org/wiki/{year}_Summer_Olympics_medal_table', index_col = 0)[2]
    dictionary[year] = df
    dictionary[f'{year}_20'] = df.head(20)
    
    print(f"Summer Olympics {year} Table".center(50), df.head(20), '_'*50, end='\n\n', sep='\n')
```
